```python
print (df.isnull().sum())
df.duplicated().sum()
df = df.dropna()
print(df)
```

```python



df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y")
df['Ship Date'] = pd.to_datetime(df['Ship Date'],  format="%d/%m/%Y")

df['Category'] = df['Category'].astype('category')
df['Region'] = df['Region'].astype('category')
df['Segment'] = df['Segment'].astype('category')


"""print(df.isnull().sum())
print(df.info())"""
print(df)
```

Transformation et Feature Engineering

```python

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.quarter
df     
```

```python
df["cost"]=df["Sales"]*0.80
df["marge"]=df["Sales"]-df["cost"]
df["taux_profit"]=(df["marge"]/df["Sales"])*100
df
```

```python
df["livrison"]=(df["Ship Date"]-df["Order Date"]).dt.days
df
```

```python
ventes_par_région=df.groupby("Region")["Sales"].sum().sort_values(ascending=False).round(2)
ventes_par_région
```
